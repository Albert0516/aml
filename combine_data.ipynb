```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('./drug_sensity_num.csv')
```

```python
## 拉取编号
df['name']=df['name'].apply(lambda x:x.split(' ')[0])
```

```python
## 敏感度量化
def sensity_to_num(des):
    if des=='不敏感':
        return 0
    elif des=='低度敏感':
        return 1
    elif des=='中度敏感':
        return 2
    else :
        return 3
df['sensity']=df['sensity'].apply(sensity_to_num)
```

```python
df.head()
```

## 抽出Y数据

```python
### 基本维度
names=df['name'].unique()
drug_names=df['drug_name'].unique()
```

```python
names.sort()
```

```python
groups=df.groupby('name')
```

```python
origin_data={}
```

```python
groups.get_group(names[2]).sort_values('drug_name').T.loc['sensity',:].values
```

```python
# 按行抽取
for name in names:
    t_row=groups.get_group(name).sort_values('drug_name').T.loc['sensity',:]
    origin_data[name]=t_row.values
```

```python
origin_data
```

```python
## 最后数据
y_data=pd.DataFrame(origin_data)
y_data=y_data.T
y_data.columns=drug_names
```

```python
y_data.head()
```

```python
y_data.to_csv('y_sensity_data.csv',index=False,encoding='utf_8_sig')
```
