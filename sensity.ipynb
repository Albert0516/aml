```python
df=pd.read_csv('./drug_sensity_num.csv')
```

```python
df['name']=df['name'].apply(lambda x:x.split(' ')[0])
```

```python
def classify(des):
    if des=='不敏感':
        return 0
    elif des=='低度敏感':
        return 1
    elif des=='中度敏感':
        return 2
    else :
        return 3
```

```python
df['sensity']=df['sensity'].apply(classify)
```
